```python
import array as arr
import numpy as np
from numpy import random


class LRUCache:
        Misses = 0
        Hits = 0
    
        def __init__(self, capacity: int):
            self.cache = []
            self.capacity = capacity

            
            
        def get(self, value):
            if value not in self.cache:
                LRUCache.Misses = LRUCache.Misses +1
                return -1
            else:
                LRUCache.Hits = LRUCache.Hits + 1
                return value
        
        def put(self, value):
            self.cache.insert(0,value)
            if len(self.cache) > self.capacity:
                self.cache.pop(-1)
                
                
cache = LRUCache(3)

cache.put(1)
print(cache.cache)
cache.put(2)
print(cache.cache)
cache.get(1)
print(cache.cache)
cache.put(3) 
print(cache.cache)
cache.get(2)
print(cache.cache)
cache.put(4)
print(cache.cache)
cache.get(1)
print(cache.cache)
cache.get(3)
print(cache.cache)
cache.put(5)
print(cache.cache)
cache.get(1)
print(cache.cache)
cache.get(2)
print(cache.cache)
cache.get(4)
print(cache.cache)
cache.get(5)
print(cache.cache)

#Ratio
m = LRUCache.Misses
h = LRUCache.Hits
print(m)
print(h)
Hit_Ratio = h/(h + m)
print(Hit_Ratio)

print('------------------\n')
print('Test2\n')
#Test andom request stream exponential dist. of length 10^5 as an input

A = np.random.exponential(1,size=100000)
A = list(np.around(np.array(A),2))
val = np.random.choice(A,5)
print(val)


cache = LRUCache(3)

cache.put(val[0])
print(cache.cache)
cache.put(val[1])
print(cache.cache)
cache.get(val[0])
print(cache.cache)
cache.put(val[2])
print(cache.cache)
cache.get(val[1])
print(cache.cache)
cache.put(val[3])
print(cache.cache)
cache.get(val[0])
print(cache.cache)
cache.get(val[2])
print(cache.cache)
cache.get(val[3])
print(cache.cache)
cache.put(val[4])
print(cache.cache)
cache.get(val[0])
print(cache.cache)
cache.get(val[1])
print(cache.cache)
cache.get(val[2])
print(cache.cache)
cache.get(val[3])
print(cache.cache)
cache.get(val[4])
print(cache.cache)

m2 = LRUCache.Misses
h2 = LRUCache.Hits
print(m2)
print(h2)
Hit_Ratio_2 = h2/(h2 + m2)
print(Hit_Ratio_2)
```
